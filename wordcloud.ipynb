```python
import numpy as np # линейная алгебра
import pandas as pd # обработка данных, чтение и запись файлов CSV
import re # модуль для работы с регулярными выражениями.
from wordcloud import WordCloud, STOPWORDS  # это класс, который позволяет создавать облака слов
import matplotlib.pyplot as plt # инструментом для создания различных видов графиков и диаграмм
```

```python
# Облако слов из вставленного теекта
data = '''
Сюда вставить текст из буфера обмена
'''
```

```python
# Облако слов из файла, его нужно загрузить
# Если слова вставлены в блоке выше то это выполнять не надо
#with open('/wordcloud.csv', 'r') as file:
#    data = file.read()
```

```python
# Удаление лишних символов с помощью регулярного выражения
data = re.sub(r'[^\w\s]', '', data)

# Разбиение содержимого файла на слова
words = data.split()

# Создание DataFrame из списка слов
df = pd.DataFrame(words, columns=['Word'])

# Добавим слова-исключения по-русски.
russian_stopwords = set(STOPWORDS)
russian_stopwords.update(['это', 'как', 'так', 'что', 'все', 'я', 'мы', 'наш', 'их', 'он', 'она', 'они', 'вы', 'свой', 'мой', 'твой', 'его', 'ее', 'ваш', 'очень', 'по', 'для', 'быть', 'еще'])

# Вывод DataFrame
#print(df)

# Функция вывода облака слов
def wordcloud_draw(data, color='black'):
    words = ' '.join(data)
    wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color=color,
                          width=3500,
                          height=2000).generate(words)
    plt.figure("wordcloud", figsize=(18, 20))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

wordcloud_draw(df['Word'])
```
