```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import re
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

```python
data = '''
Текст
'''
```

```python
# Открытие файла и чтение его содержимого
#with open('/words.txt', 'r') as file:
#    data = file.read()

# Удаление лишних символов с помощью регулярного выражения
data = re.sub(r'[^\w\s]', '', data)

# Разбиение содержимого файла на слова
words = data.split()

# Создание DataFrame из списка слов
df = pd.DataFrame(words, columns=['Word'])

# Вывод DataFrame
#print(df)

def wordcloud_draw(data, color='black'):
    words = ' '.join(data)
    wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color=color,
                          width=3500,
                          height=2000).generate(words)
    plt.figure("wordcloud", figsize=(18, 20))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

wordcloud_draw(df['Word'])
```
