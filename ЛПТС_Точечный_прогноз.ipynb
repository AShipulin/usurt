```python
import numpy as np
import matplotlib.pyplot as plt

#   (,    2 )
months = np.arange(1, 25)
sales = np.array([10, 12, 11, 13, 14, 15, 16, 18, 17, 20, 19, 21,
                  22, 23, 24, 25, 26, 28, 27, 30, 29, 31, 32, 33])

#    
coefficients = np.polyfit(months, sales, 1)
trend_line = np.poly1d(coefficients)
trend = trend_line(months)

#    6 
future_months = np.arange(25, 31)
forecast = trend_line(future_months)

# 
plt.figure(figsize=(12, 6))
plt.plot(months, sales, 'o', label=' ', markersize=5)
plt.plot(months, trend, 'r-', label=' ')
plt.plot(future_months, forecast, 'go', label=' ', markersize=8)

#  
plt.title('   ', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel('  ( .)', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

#   
print("   :")
for month, value in zip(future_months, forecast):
    print(f" {month}: {value:.2f}  .")
```
