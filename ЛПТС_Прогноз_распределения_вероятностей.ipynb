# Прогноз распределения вероятностей

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Данные для прогноза (объем продаж в млн руб.)
mean = 12.0   # среднее значение
std_dev = 0.8  # стандартное отклонение

# Генерация данных для нормального распределения
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = norm.pdf(x, mean, std_dev)

# Определение интервалов для сценариев
pessimistic = mean - std_dev
realistic = mean
optimistic = mean + std_dev

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'k-', label='Распределение вероятностей')
plt.fill_between(x, y, where=(x <= pessimistic), color='red', alpha=0.3, label='Пессимистичный (<11.2)')
plt.fill_between(x, y, where=(x >= optimistic), color='green', alpha=0.3, label='Оптимистичный (>12.8)')
plt.fill_between(x, y, where=(x > pessimistic) & (x < optimistic), color='blue', alpha=0.3, label='Реалистичный (11.2–12.8)')

# Вертикальные линии для среднего и границ
plt.axvline(mean, color='black', linestyle='--', linewidth=1, label=f'Среднее: {mean}')
plt.axvline(pessimistic, color='red', linestyle=':', linewidth=1)
plt.axvline(optimistic, color='green', linestyle=':', linewidth=1)

# Настройки графика
plt.title('Прогноз объема продаж с распределением вероятностей', fontsize=14)
plt.xlabel('Объем продаж (млн руб.)', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
